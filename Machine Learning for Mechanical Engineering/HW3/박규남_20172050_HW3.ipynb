## 2) Perspective homography for image 1 and image 2

```python
## your code here
## Construct feature matrix using homography H, and a vector having entries of matching points in image 2 
theta_1 = pos1[0]
theta_1 = [theta_1[int(i/2)] if i%2==0 else 0 for i in range(len(theta_1)*2)]
theta_1 = np.array(theta_1).reshape(-1,1)
print(theta_1.shape)
theta_2 = pos1[1]
theta_2 = [theta_2[int(i/2)] if i%2==0 else 0 for i in range(len(theta_2)*2)]
theta_2 = np.array(theta_2).reshape(-1,1)
print(theta_2.shape)
theta_3 = [1 if i%2==0 else 0 for i in range(len(pos1[0])*2)]
theta_3 = np.array(theta_3).reshape(-1,1)
print(theta_3.shape)
theta_4 = pos1[0]
theta_4 = [theta_4[int(i/2)] if i%2==1 else 0 for i in range(len(theta_4)*2)]
theta_4 = np.array(theta_4).reshape(-1,1)
print(theta_4.shape)
theta_5 = pos1[1]
theta_5 = [theta_5[int(i/2)] if i%2==1 else 0 for i in range(len(theta_5)*2)]
theta_5 = np.array(theta_5).reshape(-1,1)
print(theta_5.shape)
theta_6 = [1 if i%2==1 else 0 for i in range(len(pos1[0])*2)]
theta_6 = np.array(theta_6).reshape(-1,1)
print(theta_6.shape)
xx = np.multiply(pos1[0], pos2[0])
yx = np.multiply(pos1[1], pos2[1])
theta_7 = [-xx[int(i/2)] if i%2==0 else -yx[int(i/2)] for i in range(len(xx)*2)]
theta_7 = np.array(theta_7).reshape(-1,1)
print(theta_7.shape)
xy = np.multiply(pos2[0], pos1[1])
yy = np.multiply(pos2[1], pos1[1])
theta_8 = [-xy[int(i/2)] if i%2==0 else -yy[int(i/2)] for i in range(len(xx)*2)]
theta_8 = np.array(theta_8).reshape(-1,1)
print(theta_8.shape)
theta = np.hstack([theta_1, theta_2,theta_3, theta_4,theta_5, theta_6, theta_7, theta_8])
theta = np.asmatrix(theta)
print(theta)

b = [pos2[0][int(i/2)] if i%2==0 else pos2[1][int(i/2)] for i in range(len(pos2[0])*2)]
b = np.array(b).reshape(-1,1)
print(b.shape)
## your code here
## Define perspective_theta using linear regression

perspective_theta = (theta.T*theta).I*theta.T*b
perspective_theta = np.append(perspective_theta,[[1]],axis=0)
print(perspective_theta)
perspective_theta = perspective_theta.reshape(3,3)
perspective_theta
```
