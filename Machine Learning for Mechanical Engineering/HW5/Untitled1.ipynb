```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

%matplotlib inline
```

```python

# data generation
m = 5000
mu = np.array([0, 0])
sigma = np.array([[3, 1.5], 
                  [1.5, 1]])

X = np.random.multivariate_normal(mu, sigma, m)
print(X.shape)
X = np.asmatrix(X)

fig = plt.figure(figsize=(10, 6))
plt.plot(X[:,0], X[:,1], 'k.')
plt.axis('equal')
plt.show()
```

```python
S = 1/(m-1)*X.T*X
S = np.asmatrix(S)
 
D, V = np.linalg.eig(S)

idx = np.argsort(-D)
D = D[idx]
V = V[:,idx]

print(D)
print(V)
```
