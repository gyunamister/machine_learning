## Problem 2

### 1) Generate a random samples using *randn* command in python

```python
m = 10000
x = np.random.randn(m,1);
x
```

### 2) Count the number of samples which fall into a band

```python
count = 0
for i in x:
    if i > -2 and i < 2:
        count+=1
count
```

### 3) count the ratio of samples that fall into a band to the total number of samples

```python
ratio = count/m
ratio
```

### 4) Plot the graph to show how the ratio changes as the number of samples increase

```python
l=100
u=1000
ratio = []
for m in range(l,u):
    count=0
    x = np.random.randn(m,1)
    for i in x:
        if i > -2 and i < 2:
            count+=1
    ratio.append(count/m)
```

```python
plt.figure(figsize=(8,6))
X = np.linspace(l,u,u-l)
plt.plot(X,ratio)
plt.show()
```

## Problem 3 - Monty Hall Problem with Monte Carlo

```python
from random import shuffle, choice
```

```python
doors = ['car', 'goat', 'goat']
print(doors)
shuffle(doors)
print(doors)
```

```python


def simulation(num_trial, strategy):
    if strategy == 'stay':
        stay_count=0
    if strategy == 'switch':
        switch_count=0
    for i in range(num_trial):
        doors = ['car', 'goat', 'goat']
        shuffle(doors)
        c = choice([0,1,2])
        if strategy == 'stay':
            if doors[c] == 'car':
                stay_count+=1
        elif strategy=='switch':
            pool = [0,1,2]
            for index, x in enumerate(doors):
                if index == c:
                    continue
                if x == 'goat':
                    c_dummy = index
            pool.remove(c)
            pool.remove(c_dummy)
            final_c = pool[0]
            if doors[final_c] == 'car':
                switch_count+=1
    if strategy == 'stay':
        print("staying: {}".format(stay_count/num_trial*100))
    if strategy == 'switch':
        print("switching: {}".format(switch_count/num_trial*100))
```

```python
num_trial = 10000
print("start simulation")
simulation(num_trial, 'stay')
simulation(num_trial, 'switch')
```

## Problem 4

```python
from six.moves import cPickle

x_reg = cPickle.load(open('./problem4/x_reg.pkl', 'rb'))
y_reg = cPickle.load(open('./problem4/y_reg.pkl', 'rb'))
print(x_reg.shape)
```

### 1) Linear regression

```python
x_reg = x_reg.reshape(-1,1)
y_reg = y_reg.reshape(-1,1)
A = np.hstack([x_reg, x_reg**0])
A = np.asmatrix(A)

theta = (A.T*A).I*A.T*y_reg

print('theta:\n', theta)
```

### 2) PCA

```python
m = y_reg.shape[0]
X = np.hstack([x_reg, y_reg])
X = np.asmatrix(X)

S = 1/(m-1)*X.T*X
S = np.asmatrix(S)
 
D, V = np.linalg.eig(S)

idx = np.argsort(-D)
D = D[idx]
V = V[:,idx]

print(D)
print(V)
```

```python
# to plot
plt.figure(figsize=(40, 20))
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.plot(x_reg, y_reg, 'ko', label="data")
plt.axis('equal')

# to plot a linear regression line (fitted line)
xp = np.arange(-5, 5, 0.1).reshape(-1, 1)
yp = theta[0,0]*xp + theta[1,0]

plt.plot(xp, yp, 'r', linewidth=2, label="Linear Regression")

# to plot PCA
h = V[1,0]/V[0,0]
xp = np.arange(-5, 5, 0.1).reshape(-1, 1)
yp = h*xp
plt.plot(xp,yp, 'b', linewidth=2, label='PCA')

plt.legend(fontsize=15)
plt.axis('scaled')
plt.show()

```

### 3) Comment on why the results are different
Linear regression minimizes the sum of squares error(square of predicted y and y). On the other hand, PCA minimizes the sum of orthogonal distance between the line and data points.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n0 = 200
n1 = 200

sigma = [[19, -4],
         [-4, 1]]

x0 = np.random.multivariate_normal([0.7,0.7], sigma, n0).T        # data in class 0
x1 = np.random.multivariate_normal([-0.5,-0.5], sigma, n1).T      # data in class 0

print(x0.shape)
x0 = np.asmatrix(x0)
x1 = np.asmatrix(x1)

plt.figure(figsize = (10, 6))
plt.plot(x0[:,0],x0[:,1],'r.')
plt.plot(x1[:,0],x1[:,1],'b.')
plt.ylim([-8, 8])
plt.xlim([-14, 14])
plt.show()
```

```python
mu0 = np.mean(x0, axis=1)
mu1 = np.mean(x1, axis=1)

S0 = 1/(n0-1) * (x0-mu0)*(x0-mu0).T
S1 = 1/(n1-1) * (x1-mu1)*(x1-mu1).T

w = (n0*S0 + n1*S1).I * (mu0-mu1)
print(w)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x0[0,:],x0[1,:],'r.')
plt.plot(x1[0,:],x1[1,:],'b.')

xp = np.arange(-4, 6, 0.1)
yp = w[1,0]/w[0,0]  * xp
plt.plot(xp, yp, 'k')
plt.axis('equal')
plt.ylim([-2, 6])
plt.show()
```

### 2)

```python
X = np.hstack([x0.T,x1.T])
X = np.asmatrix(X)
print(X.shape)
S = 1/(m-1)*X.T*X
S = np.asmatrix(S)
 
D, V = np.linalg.eig(S)

idx = np.argsort(-D)
D = D[idx]
V = V[:,idx]

print(D)
print(V)
```

```python
h = V[1,0]/V[0,0]
xp = np.arange(-6, 6, 0.1)
yp = h*xp

fig = plt.figure(figsize=(10, 6))
plt.plot(X[:,0], X[:,1], 'k.')
plt.plot(xp, yp, 'r', linewidth=4.0)
plt.axis('equal')
plt.show()
```
