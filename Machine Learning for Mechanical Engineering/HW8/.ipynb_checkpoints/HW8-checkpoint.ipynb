### Problem7
Write a python code to find a linear transformation  y=Ax such that  y are uncorrelated (or independent), and plot the data after transformation.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mu = np.array([0, 0])
sigma = 1./2.*np.array([[5, -3], [-3, 5]])

m = 1000
x = np.random.multivariate_normal(mu, sigma, m)

plt.figure(figsize=(10,6))
plt.plot(x[:,0], x[:,1], '.')
plt.axis('equal')
plt.ylim([-8,8])
plt.show()
```

```python
x
```

```python
import math
A = np.array([[-math.sqrt(2)/2,-math.sqrt(2)/2],[-math.sqrt(2)/2,math.sqrt(2)/2]])
A = np.asmatrix(A)
A
```

```python
ys=[]
for row in x:
    row = row.reshape(-1,1)
    y = A*row
    ys.append(y)
ys = np.array(ys)
ys = ys.reshape(-1,2)
ys
```

```python
plt.figure(figsize=(10,6))
plt.plot(ys[:,0], ys[:,1], '.')
plt.axis('equal')
plt.ylim([-8,8])
plt.show()
```
